# LEARN SEABORN PYTHON LIBRARY

**seaborn line plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,2,7,6]

x_1=pd.DataFrame({"var":var,"var_1":var_1})

sns.lineplot(x="var",y="var_1",data=x_1)

plt.show()
```

**we can plot without dataframe but label will not be showed in that case and matplotlib type graph shown**

```python
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,2,7,6]
sns.lineplot(x=var,y=var_1)

plt.show()
```

**now we use real type data set file to work on it through seaborn and plot line graph between two data**

```python
y_1=sns.load_dataset("penguins")
```

```python
y_1
```

**"Hue parameter": used to plot graph between two with respect to popular data in dataset automatically, as mentioned below sex is popular data here**

**"palette parameter": used to change the color of plot**

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",
             data=y_1,hue="sex",style="sex",palette="plasma",
             markers=["o",">"])
plt.show()
```

**we can use dashes parameter to remove dotted line by giving it false OR "In seaborn legend() function is used to remove legend of plot but if not used then it has always legend on graph"**

**now we used grid ot title parameter**

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",
             data=y_1,hue="sex",style="sex",palette="plasma",
             markers=["o",">"],dashes=False)

plt.title("penguin",fontsize=18)
plt.grid()
plt.show()
```

# PLOTTING BAR CHART IN SEABORN

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df=sns.load_dataset("penguins")
df
```

```python
sns.barplot(x="species",y="bill_length_mm",data=df,
            hue="sex")
plt.show()
```

**change the graph by any order or changing the sex data plot order according to hue_order parameter**

```python
order_1=["Chinstrap","Adelie","Gentoo"]
sns.barplot(x="species",y="bill_length_mm",data=df,
            hue="sex",order=order_1,hue_order=["Female","Male"],
            )
plt.show()
```

**for orientation to be horizontal so only apply on numeric data not on string data column**

```python

sns.barplot(x="bill_depth_mm",y="bill_length_mm",
            data=df,orient="h")
plt.show()
```

```python
sns.barplot(x="species",y="bill_length_mm",
            data=df,color="g")
plt.show()
```

we can change the saturation also with parameter


```python
sns.barplot(x="species",y="bill_length_mm",data=df,
            hue="sex",palette="ocean",saturation=1,
            err_kws={'color': 'b'},
            capsize=1)
plt.show()
```

**by using sns.set(style="darkgrid") grid will be removed**

```python
sns.barplot(x="species",y="bill_length_mm",data=df,
            hue="sex",palette="ocean",
            err_kws={'color': 'b','linewidth':7},alpha=1)
plt.show()
```

# HEATMAP METHOD OF PLOTTING IN SEABORN

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=np.linspace(1,10,20).reshape(4,5)
df
```

```python
sns.heatmap(df)
plt.show()
```

```python
dff=sns.load_dataset("anagrams")
x=dff.drop(columns=["attnr"],axis=1).head(15)
x
```

```python
sns.heatmap(x)
plt.show()
```

**some parameters as shown**

**annot parameter used to show the value on heatmap**

```python
df=np.linspace(1,10,10).reshape(2,5)
df
sns.heatmap(df,vmin=0,vmax=20,cmap="gist_gray",annot=True)
plt.show()
```

**using user input in place of annot parameter**

```python
ar=np.array([["a0","a1","a2","a3","a4"],
             ["b0","b1","b2","b3","b4"]])
ar
```

```python
y={"fontsize":15,"color":"r"}
sns.heatmap(df,vmin=0,vmax=20,cmap="gist_gray",
            annot=ar,fmt="s",annot_kws=y,linewidth=5,linecolor="black")
plt.show()
```

**to remove color map or label and set lable using set parameter**

```python
y={"fontsize":15,"color":"r"}
v=sns.heatmap(df,vmin=0,vmax=20,cmap="gist_gray",
            annot=ar,fmt="s",annot_kws=y,linewidth=5,linecolor="black",
            cbar=False,xticklabels=False,yticklabels=False)

v.set(xlabel="python",ylabel="heatgraph")
plt.show()
```

## VIOLIN PLOT METHOD IN SEABORN

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

pf=sns.load_dataset("tips")
pf
```

```python
sns.violinplot(x="day",y="total_bill",data=pf)
plt.show()
```

# pair plot method 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

pf=sns.load_dataset("tips").head(10)
pf
```

```python
sns.pairplot(pf)
plt.show()
```

```python
sns.pairplot(pf,hue="sex")
plt.show()
```

```python
sns.pairplot(pf,vars=["total_bill","tip"],hue="sex",
             hue_order=["Female","Male"],palette="rocket")
plt.show()
```

```python
sns.pairplot(pf,hue="sex",hue_order=["Female","Male"],y_vars=["total_bill","tip"],palette="rocket")
plt.show()
```

**we can change the kind of plot in pair method like using kind parameter we have forms lke kde,scatter,hist,reg**

diag_kind parameter used to fix the plot not showiing using kind so t changes accordingly below kde to hist is shown

```python
sns.pairplot(pf,hue="sex",hue_order=["Female","Male"],kind="kde",diag_kind="hist")
plt.show()
```

```python
sns.pairplot(pf,hue="sex",hue_order=["Female","Male"],markers=["*","<"])
plt.show()
```

# BOX PLOT METHOD

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

pf=sns.load_dataset("tips")
pf
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=pf)
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=pf,hue="time",palette='dark:g',
            order=["Fri","Sun","Sat","Thur"])
plt.show()
```

to show mean in box plot

there are some more parameters like linewidth, orient (for orient both data should be numerical) etc 

```python
sns.boxplot(x="day",y="total_bill",data=pf,hue="time",palette='gnuplot2',
            order=["Fri","Sun","Sat","Thur"],showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"magenta"})
plt.show()
```

FOR SINGLE DATASET PLOT

```python
sns.set(style="white")

sns.boxplot(y=pf["total_bill"])
plt.show()
```

# styling plot while plotting in seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

pf=sns.load_dataset("tips")
pf
```

```python
sns.barplot(x="day",y="total_bill",data=pf)
plt.show()
```

making of grid using seaborn


```python
sns.set_style("darkgrid")

sns.barplot(x="day",y="total_bill",data=pf)

plt.show()
```

to remove axis line we use despine parameter

```python
sns.set_style("white")

sns.barplot(x="day",y="total_bill",data=pf)

sns.despine()
plt.show()
```

to increase figure size it is the funtion of matplotlib by using plt.figure()

```python
sns.set_style("white")
plt.figure(figsize=(7,8))

sns.barplot(x="day",y="total_bill",data=pf)


plt.show()
```

change in content like fonts etc. by using sns.set_context() parameter in which there is many options like paper,poster,talk etc

```python
sns.set_style("white")

sns.set_context("paper",font_scale=1)

sns.barplot(x="day",y="total_bill",data=pf)


plt.show()
```

```python
sns.set_style("white")

sns.set_context("poster",font_scale=1)

sns.barplot(x="day",y="total_bill",data=pf,palette="cool")


plt.show()
```

# MULTIPLE PLOTS (FACET-GRID) IN SEABORN

```python
import seaborn as sns
import matplotlib.pyplot as plt

pf=sns.load_dataset("tips")
pf
```

```python
plt.figure(figsize=(10,5))
graph=sns.FacetGrid(pf,col="sex",hue="day")

graph.map(plt.scatter,'total_bill','tip').add_legend()

plt.show()
```

```python

graph=sns.FacetGrid(pf,col="sex",hue="day",palette="rocket")

graph.map(plt.bar,'total_bill','tip',edgecolor="b").add_legend()

plt.show()
```
